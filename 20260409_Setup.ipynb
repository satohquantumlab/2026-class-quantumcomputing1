## Step 2: 最初の量子回路を作る
量子回路は、量子ビットに対する操作を順番に記述したものです。  
まずは **1量子ビットの空の回路** を作って表示してみましょう。

```python
from qiskit import QuantumCircuit

# 1量子ビット、1古典ビットの回路を作成
qc = QuantumCircuit(1, 1)

# 回路を表示（まだ何も操作していないので空っぽ）
qc.draw(output='mpl')
```

---
## Step 3: 量子ゲートを1つ追加してみる
**Hゲート（アダマールゲート）** は量子コンピューティングで最も重要なゲートの1つです。  
|0⟩ 状態を |0⟩と|1⟩の **重ね合わせ状態** に変換します。  

詳しくは次回 (#2) で学びますが、今日は「動く」ことだけ確認しましょう。

```python
# Hゲートを追加
qc.h(0)

# 測定を追加（量子ビットの状態を読み出す）
qc.measure(0, 0)

# 回路を表示
qc.draw(output='mpl')
```

---
## Step 4: シミュレータで実行する
この回路を古典コンピュータ上のシミュレータで1024回実行し、  
測定結果の統計を見てみましょう。

```python
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram

# シミュレータで実行
simulator = AerSimulator()
result = simulator.run(qc, shots=1024).result()
counts = result.get_counts()

print(f"測定結果: {counts}")
plot_histogram(counts)
```

**期待される結果：** `{'0': ~512, '1': ~512}` のように、0と1がほぼ半々で出るはずです。  
これが **重ね合わせ** の効果です。コインを投げるように、測定するまでどちらになるか分かりません。  

次回 (#2) では、この「重ね合わせ」と「量子もつれ」について詳しく学びます。

---
## Step 5: 2量子ビットでベル状態を覗いてみる（おまけ）
次回の予習として、2つの量子ビットを「もつれ」させてみましょう。  
結果を見て、何か不思議なことに気づきますか？

```python
# 2量子ビットの回路
bell = QuantumCircuit(2, 2)
bell.h(0)       # 1つ目の量子ビットにHゲート
bell.cx(0, 1)   # 1つ目で2つ目を制御するCXゲート（次回詳しく学びます）
bell.measure([0, 1], [0, 1])

bell.draw(output='mpl')
```

```python
# 実行して結果を見る
result = simulator.run(bell, shots=1024).result()
counts = result.get_counts()

print(f"測定結果: {counts}")
plot_histogram(counts)
```

**観察：** `00` と `11` しか出ない（`01` や `10` は出ない）ことに注目してください。  
2つの量子ビットが「もつれて」おり、片方が0なら他方も必ず0、片方が1なら他方も必ず1になります。  

**これは偶然ではありません。** この不思議な相関の正体を、#2と#5で詳しく探ります。
